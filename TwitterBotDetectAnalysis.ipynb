```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys, codecs, json
from twython import TwythonStreamer, Twython
from datetime import datetime
from time import time

# Load in OAuth Tokens
app_key = ""
app_sec = ""
user_key = ""
user_sec = ""

api = Twython(app_key, app_sec, user_key, user_sec)
api.verify_credentials()
tweets_data = api.search(q='ny')
tweets_data = tweets_data['statuses']
```

```python
data = pd.DataFrame(data=[tweet['user']['name'] for tweet in tweets_data], columns=['User_Name'])
data['User_ID'] =np.array([tweet['user']['id'] for tweet in tweets_data])
data['User_Screen_name'] =np.array([tweet['user']['screen_name'] for tweet in tweets_data])
data['User_Loc'] =np.array([tweet['user']['location'] for tweet in tweets_data])
data['User_Lang'] =np.array([tweet['user']['lang'] for tweet in tweets_data])
data['User_Following'] =np.array([tweet['user']['friends_count'] for tweet in tweets_data])
data['User_Followers'] =np.array([tweet['user']['followers_count'] for tweet in tweets_data])
data['User_List'] =np.array([tweet['user']['listed_count'] for tweet in tweets_data])
data['User_Liked_Tweets'] =np.array([tweet['user']['favourites_count'] for tweet in tweets_data])
data['User_Status'] =np.array([tweet['user']['statuses_count'] for tweet in tweets_data])
data['User_Creation'] =np.array([tweet['user']['created_at'] for tweet in tweets_data])
data['Tweet'] =np.array([tweet['text'] for tweet in tweets_data])
data['Time'] =np.array([tweet['created_at'] for tweet in tweets_data])
import time
data['Create_Year'] = pd.DatetimeIndex(data['User_Creation']).year
```

```python
import re
ids = api.get_followers_ids(id='970382127840755713')
idlist = ids['ids']
loclist = []
results = list(map(str, idlist))
```

```python
for i in idlist:
    res = api.lookup_user(user_id=i)
    for loc in res:
        x = re.sub('[^a-zA-Z,]+', ' ', loc['location'])
        x = x.replace(',', '')
        loclist.append(x)
```

```python
loc2=[]
for i in loclist:
    i = i.replace(',', '').replace('.', '').strip()
    shortword = re.compile(r'\W*\b\w{1,3}\b')
    text = shortword.sub('', i)
    loc2.append(text)
loc2 = list(filter(None, loc2))
loc2

```

```python
follower_loc = pd.DataFrame(data=loc2, columns=['Follower_Location'])
follower_loc
```
